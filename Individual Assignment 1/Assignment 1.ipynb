# Plot 1 (Macro): Increase in the percentage of the global population affected by depression over the past 30 years 

```python
df = pd.read_csv(r"C:\Users\yuant\OneDrive\Documents\DSA4262\Assignment 1\IHME-GBD_2023_DATA-b6e525ad-1\IHME-GBD_2023_DATA-b6e525ad-1.csv")
df.head()
```

```python
df = df.loc[:, ['measure', 'metric', 'year', 'val', 'upper', 'lower']]
df.head()
```

```python
df.info()
```

```python
df_prev_perc = df[(df['measure'] == 'Prevalence') & (df['metric'] == 'Percent')]
df_prev_perc['val'] = df_prev_perc['val'] * 100
df_prev_perc['upper'] = df_prev_perc['upper'] * 100
df_prev_perc['lower'] = df_prev_perc['lower'] * 100

df_prev = df[(df['measure'] == 'Prevalence') & (df['metric'] == 'Rate')]
```

```python
df_prev_perc.head()
```

```python
band = alt.Chart(df_prev_perc).mark_area(opacity=0.3).encode(
    x=alt.X(
        'year:Q',
        axis=alt.Axis(
            format='d', 
            title='Year',
            values=list(range(1993, 2024, 2)) 
        ),
        scale=alt.Scale(domain=[1992, 2024])
    ),
    y=alt.Y('lower:Q', axis=alt.Axis(title='Percentage (%)')), 
    y2='upper:Q'
)

line = alt.Chart(df_prev_perc).mark_line(point=True).encode(
    x=alt.X(
        'year:Q',
        axis=alt.Axis(
            format='d', 
            title='Year',
            values=list(range(1993, 2024, 2))
        ),
        scale=alt.Scale(domain=[1992, 2024])
    ),
    y='val:Q'
)


plot1 = (band + line).properties(
    width=850,
    height=500,
    title='Increase in Percentage of Global Population with Depression Over Time (1993â€“2023)'
).configure_axisX(
    labelFontSize=14,
    labelAngle=0,
    titleFontSize=18,
    titlePadding=15
).configure_axisY(
    labelFontSize=14,
    titleFontSize=18,
    titlePadding=15
).configure_title(
    fontSize=22
)

plot1

```

## Higher Internet Addiction Scores in Individuals with Severe and Extremely Severe Depression Levels

```python
df2 = pd.read_excel(r"C:\Users\yuant\OneDrive\Documents\DSA4262\Assignment 1\Internet_addiction.xlsx")
df2.head()
```

```python
df2.info()
```

```python
df2["Depression_Level"].unique()

df2["Self-Esteem_Level"].unique()

df2["Anxiety_Level"].unique()

df2["Gender"].unique()
```

Some of these categories have repeated values that have not been properly formatted. I will reformat them below:

```python
df2['Depression_Level'] = df2['Depression_Level'].str.strip()

df2['Depression_Level'] = (
    df2['Depression_Level']
      .str.strip()
      .replace({'Extremely': 'Extremely Severe'})
)

df2['Self-Esteem_Level'] = df2['Self-Esteem_Level'].str.strip()

df2['Internet_Addiction_Level'] = df2['Internet_Addiction_Level'].str.strip()
```

```python
'''
df2.groupby('Age')['Age'] \
  .agg(
      count='count'
  )
'''
```

Let's have a look at the Internet Addiction Level ranges since it was not specified along with the dataset. 

```python
df2.groupby('Internet_Addiction_Level')['Internet_Addiction_Score'] \
  .agg(
      min_score='min',
      max_score='max',
      mean_score='mean',
      count='count'
  )
```

```python
df2.groupby("Depression_Level")[[
    "Internet_Addiction_Score",
    "Self-Esteem_Score",
    "Depression_Score",
    "Anxiety_Score",
    "Stress_Score",
    "hours_per_day"
]].mean()
```

```python
depr_order = ['Normal', 'Mild', 'Moderate', 'Severe', 'Extremely Severe']

color_list = ['#c6dbef', '#9ecae1', '#6baed6', '#3182bd', '#08519c']

box = alt.Chart(df2).mark_boxplot(
    extent='min-max',
    size=55
).encode(
    alt.X(
        'Depression_Level:N',
        sort=depr_order,
        title='Depression Level'
    ),
    alt.Y(
        'Internet_Addiction_Score:Q',
        title='Internet Addiction Score'
    ),
    alt.Color(
        'Depression_Level:N',
        scale=alt.Scale(domain=depr_order, range=color_list),
        legend=None
    )
)


line = alt.Chart(pd.DataFrame({'y':[50]})).mark_rule(
    color='red',   
    strokeWidth=2,
    strokeDash=[5,5]
).encode(
    y='y:Q'
)


label = alt.Chart(pd.DataFrame({'y':[50]})).mark_text(
    text='Problematic / Addicted',
    color='red',
    fontWeight='bold',
    align='left',
    baseline='bottom',
    dx=5, 
    dy=-5 
).encode(
    y='y:Q',
    x=alt.value(10)
)


box2 = (box + line + label).properties(
    title=alt.TitleParams(
        text="Higher Internet Addiction Scores in Individuals",
        subtitle="with Severe and Extremely Severe Depression Levels",
        fontSize=24,          
        subtitleFontSize=24,  
        subtitleFontWeight='bold', 
        anchor='middle',       
        offset=20                 
    ),
    width=700,
    height=500
).configure_axisX(
    labelFontSize=18,
    labelAngle=0,
    titleFontSize=20,
    titlePadding=25
).configure_axisY(
    labelFontSize=15,
    titleFontSize=18
)

box2
```

# Plot 3 (Actionable): Top Social Media Platforms Used by Singaporean Young Adults

```python
df3 = pd.DataFrame({
    "Platform": [
        "WhatsApp",
        "YouTube",
        "Instagram",
        "Facebook",
        "Telegram",
        "TikTok",
        "Discord",
        "Twitter",
        "WeChat",
        "Xiaohongshu",
        "Twitch",
        "Snapchat",
        "Weibo",
        "Others",
        "Don't Use"
    ],
    "Platforms Used (%)": [91.1, 80.2, 88.6, 58.7, 73.2, 58.6, 30.1, 26.5, 15.5, 16.0, 10.9, 6.1, 3.5, 2.1, 0.7]
})

df3.head()
```

```python
platform_type_map = {
    "WhatsApp": "Messaging",
    "Telegram": "Messaging",
    "WeChat": "Messaging",
    "Discord": "Messaging",

    "YouTube": "Media Sharing",
    "TikTok": "Media Sharing",
    "Twitch": "Media Sharing",
    "Xiaohongshu": "Media Sharing",
    "Snapchat": "Media Sharing",

    "Instagram": "Networking",
    "Facebook": "Networking",

    "Twitter": "Microblogging",
    "Weibo": "Microblogging",

    "Others": "Others",
    "Don't Use": "Others"
}

df3["Platform_Type"] = df3["Platform"].map(platform_type_map)
df3.head()
```

```python
bar3 = alt.Chart(df3).transform_window(
    rank='rank(Platforms Used (%))',
    sort=[alt.SortField('Platforms Used (%)', order='descending')]
).transform_filter(
    alt.datum.rank <= 10
).mark_bar().encode(
    x=alt.X(
        "Platform:N",
        sort=alt.SortField(field="Platforms Used (%)", order="descending"),
        title="Platform"
    ),
    y=alt.Y(
        "Platforms Used (%):Q",
        title="Respondents Using Platform (%)"
    ),
    color=alt.Color(
        "Platform_Type:N",
        title="Platform Type",
        legend=alt.Legend(
            titleFontSize=16,
            labelFontSize=14,
            symbolSize=200
        )
    ),
    tooltip=[
        "Platform:N",
        "Platforms Used (%):Q",
        "Platform_Type:N"
    ]
).properties(
    width=700,
    height=400,
    title="Top 10 Social Media Platforms Used By Young Adults in Singapore"
).configure_axisX(
    labelFontSize=14,
    labelAngle=0,
    titleFontSize=18,
    titlePadding=15
).configure_axisY(
    labelFontSize=14,
    titleFontSize=18,
    titlePadding=15
).configure_title(
    fontSize=22
)

bar3
```
